```python
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import math
```

```python
iris = sns.load_dataset('iris')
sns.load_dataset('iris')
# Continuous = everything minus species
```

```python
# What does the distribution of petal lengths look like?
sns.histplot(data=iris, x="petal_length", kde=True)
plt.xlabel("Petal Length (cm)")
plt.ylabel("Frequency")
plt.title("Distribution of Petal Lengths in Iris Dataset")
plt.show()
```

```python
sns.scatterplot(data=iris, x="petal_length", y="petal_width")
plt.xlabel("Petal Length (cm)")
plt.ylabel("Petal Width (cm)")
plt.title("Relationship Between Petal Length and Petal Width")
plt.show()
```

```python
# Would it be reasonable to predict species based on sepal width and sepal length? For this, you'll visualize two numeric columns through the lense of a categorical column.

sns.scatterplot(data=iris, x="sepal_length", y="sepal_width", hue="species", palette="Set1")
plt.xlabel("Sepal Length (cm)")
plt.ylabel("Sepal Width (cm)")
plt.title("Sepal Length vs Sepal Width by Species")
plt.legend(title="Species")
plt.show()
```

```python
sns.pairplot(iris, hue="species", markers=["o", "s", "D"])
plt.show()
corner = True
correlation_matrix = iris.corr()
correlation_with_species = correlation_matrix["species"].abs().sort_values(ascending=False)
print(correlation_with_species)
```

```python
# Load the anscombe dataset from seaborn. Use pandas to group the data by the dataset column, and calculate summary statistics for each dataset. What do you notice?

import seaborn as sns
import pandas as pd

# Load the Anscombe dataset
anscombe = sns.load_dataset("anscombe")

# Group the data by the "dataset" column
grouped_data = anscombe.groupby("dataset")

# Calculate summary statistics for each dataset
summary_statistics = grouped_data.describe()

# Print the summary statistics
print(summary_statistics)
```

```python
swiss = data("swiss")
```

```python
# threshold = 50 this is a variable method, just using a number works
swiss['is_catholic'] = swiss['Catholic'] > 50
swiss.head()
```

```python
# Create an attribute named is_catholic that holds a boolean value of whether or not the province is Catholic. (Choose a cutoff point for what constitutes catholic)
# Choose a threshold (e.g., 60%) to define what constitutes a Catholic province
threshold = 60

# Create the is_catholic attribute as a boolean column
swiss['is_catholic'] = swiss['Catholic'] > threshold

# Visualize the dataset with the is_catholic attribute
print(swiss.head())
```

```python
# Does whether or not a province is Catholic influence fertility?
# Calculate the correlation coefficient between Catholic and Fertility
correlation = swiss['Catholic'].corr(swiss['Fertility'])
print(f"Correlation between Catholic and Fertility: {correlation:.2f}")
```

```python
sns.boxplot(data=swiss, x='is_catholic', y='Fertility')
```

```python
swiss.corr()
```

```python
sns.heatmap(swiss.corr()[:1], cmap='Greens', annot=True)
```

```python
sns.scatterplot(data=swiss, x='Catholic', y='Fertility')
plt.xlabel("Percentage of Catholics")
plt.ylabel("Fertility Rate")
plt.title("Relationship between Catholic Population and Fertility")
plt.show()
```

```python
# Calculate the correlation coefficients between Fertility and other measures
correlation_with_fertility = swiss.corr()['Fertility'].drop('Fertility')

# Find the measure that correlates most strongly with Fertility
strongest_correlation_measure = correlation_with_fertility.idxmax()
strongest_correlation_coefficient = correlation_with_fertility.max()

print(f"The measure that correlates most strongly with Fertility is '{strongest_correlation_measure}' with a correlation coefficient of {strongest_correlation_coefficient:.2f}")
```
