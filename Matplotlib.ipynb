Create and label 4 separate charts with the following equations (choose a range for x that makes sense):

y=âˆšx
y=x**3
y=2**x
y=1/(x+1)
You can use functions from the math module to help implement some of the equations above.

Combine the figures you created in the last step into one large figure with 4 subplots.

```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

```python
# Generate x values
# x = np.linspace(-10, 10, 400)
x = np.array(list(range(0,10)))

# Calculate corresponding y values using the equation
y = x**2 - x + 2

# Create the plot
plt.plot(x, y, label='y = x^2 - x + 2',c='goldenrod')

# Annotate the origin point (0, 0)
plt.annotate('Origin (0, 0)', xy=(0, 0), xytext=(1, 50), textcoords='offset points',
             arrowprops=dict(arrowstyle='->', lw=1.5), fontsize=10, color='red')

# Add labels and a legend
plt.xlabel('x')
plt.ylabel('y')
plt.title('Plot of y = x^2 - x + 2')
plt.legend()

# Show the plot
plt.grid(True)
plt.show()
```

```python
x = np.array(list(range(0,10)))

y1 = np.sqrt(x)
plt.figure(1)
plt.plot(x, y1, label='y = âˆšx',c='goldenrod')
plt.xlabel('x')
plt.ylabel('y')
plt.title('Chart 1: y = âˆšx')
plt.legend()
plt.grid(True)

y2 = x**3
plt.figure(2)
plt.plot(x,y2,label='y=x^3',c='crimson')
plt.xlabel('x')
plt.ylabel('y')
plt.title('Chart 2: y=x^3')
plt.legend()
plt.grid(True)

y3 = 2**x
plt.figure(3)
plt.plot(x,y3,label='y=2^x',c='azure')
plt.xlabel('x')
plt.ylabel('y')
plt.title('Chart 3: y=2^x')
plt.legend()
plt.grid(True)

y4 = 1/(x+1)
plt.figure(4)
plt.plot(x,y4,label='y=1/(x+1)',c='forestgreen')
plt.xlabel('x')
plt.ylabel('y')
plt.title('Chart 4: y=1/(x+1)')
plt.legend()
plt.grid(True)
```

```python
# Combine the figures you created in the last step into one large figure with 4 subplots.
plt.subplot(221)
plt.plot(x, y1)
plt.title('Chart 1: y = âˆšx')

plt.subplot(222)
plt.plot(x, y2)
plt.title('Chart 2: y = x^3')

plt.subplot(223)
plt.plot(x, y3)
plt.title('Chart 3: y = 2^x')

plt.subplot(224)
plt.plot(x, y4)
plt.title('Chart 4: y = 1/(x+1)')

plt.suptitle('four subplots')
plt.tight_layout(pad=1)
plt.show()
```
